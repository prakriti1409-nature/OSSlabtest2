```python
from sklearn.datasets import load_diabetes
import pandas as pd

data = load_diabetes()

df = pd.DataFrame(data.data, columns=data.feature_names)

df['target'] = data.target
```

```python
import matplotlib.pyplot as plt

df.drop(columns='target').hist(bins=20, figsize=(20, 15), edgecolor='red')

plt.show()
```

```python
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns
print(f'Numerical Features: {len(numerical_features)}')



print(f'Classes (target): Continuous values, not discrete')
```

```python
import seaborn as sns

correlation_matrix = df.drop(columns='target').corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np

median_target = np.median(df['target'])
df['target_binary'] = (df['target'] > median_target).astype(int)

X = df.drop(columns=['target', 'target_binary'])
Y = df['target_binary']

splits = [(0.5, 0.5), (0.7, 0.3), (0.9, 0.1)]
results_rf = []
results_svm = []

for train_size, test_size in splits:
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=train_size, test_size=test_size, random_state=38)
    
    rm = RandomForestClassifier(random_state=38)
    rm.fit(X_train, Y_train)
    rf_preds = rm.predict(X_test)
    rf_accuracy = accuracy_score(Y_test, rf_preds)
    results_rf.append(rf_accuracy)
    
    svm_model = SVC(random_state=38)
    svm_model.fit(X_train, Y_train)
    svm_preds = svm_model.predict(X_test)
    svm_accuracy = accuracy_score(Y_test, svm_preds)
    results_svm.append(svm_accuracy)

mean_rf_accuracy = np.mean(results_rf)
mean_svm_accuracy = np.mean(results_svm)

print(f' Random Forest Accuracy: {mean_rf_accuracy}')
print(f'SVM Accuracy: {mean_svm_accuracy}')
```

```python
splits_str = ['50-50', '70-30', '90-10']

plt.figure(figsize=(12, 6))

plt.plot(splits_str, results_rf, label='Random Forest', marker='o')

plt.plot(splits_str, results_svm, label='SVM', marker='o')

plt.xlabel('Train-Test ')
plt.ylabel('Ac')
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np

median_target = np.median(df['target'])
df['target_binary'] = (df['target'] > median_target).astype(int)

X = df.drop(columns=['target', 'target_binary'])
y = df['target_binary']

splits = [(0.7, 0.3), (0.8, 0.2), (0.9, 0.1)]
results_rf = []
results_svm = []

for train_size, test_size in splits:
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=train_size, test_size=test_size, random_state=42)
    
    rm = RandomForestClassifier(random_state=42)
    rm.fit(X_train, y_train)
    rf_preds = rm.predict(X_test)
    rf_accuracy = accuracy_score(y_test, rf_preds)
    results_rf.append(rf_accuracy)
    
    svm_model = SVC(random_state=42)
    svm_model.fit(X_train, y_train)
    svm_preds = svm_model.predict(X_test)
    svm_accuracy = accuracy_score(y_test, svm_preds)
    results_svm.append(svm_accuracy)

mean_rf_accuracy = np.mean(results_rf)
mean_svm_accuracy = np.mean(results_svm)

print(f' Random Forest Accuracy: {mean_rf_accuracy}')
print(f'SVM Accuracy: {mean_svm_accuracy}')
```

```python
splits_str = ['70-30', '80-20', '90-10']

plt.figure(figsize=(10, 6))

plt.plot(splits_str, results_rf, label='Random Forest', marker='o')

plt.plot(splits_str, results_svm, label='SVM', marker='o')

plt.xlabel('Train-Test ')
plt.ylabel('Ac')
plt.legend()
plt.show()
```
